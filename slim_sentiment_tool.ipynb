```python
from llmware.models import ModelCatalog
from llmware.prompts import Prompt
import swifter
import pandas as pd
```

```python

def classify_sentiment(text):
  sentiment_model = ModelCatalog().load_model("slim-sentiment-tool")
  response_sentiment = sentiment_model.function_call(text, get_logits=False)
  sentiment_output = response_sentiment['llm_response']['sentiment'][0]
  return sentiment_output
```

```python
sentiment_data=pd.read_csv('sentiment_sample_data.csv')
sentiment_data=sentiment_data.head(10)
```

```python
sentiment_data['sentiment'] = sentiment_data['reviewText'].swifter.apply(classify_sentiment)
```
